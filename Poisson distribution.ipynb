Poisson distribution,belli bir sabit zaman birim aralığında meydana gelme sayısının olasığını ifade eder. Bu zaman aralığında ortalama olay meydana gelme sayısının bilindiği ve herhangi bir olayla onu hemen takip eden olay arasındaki zaman farkının, önceki zaman farklarından bağımsız oluştuğu kabul edilir.

```python
import sympy as sym
from sympy import Symbol
from sympy import pprint
import sympy.plotting as syp
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook
```

```python
x=Symbol('x')
k=Symbol('lambda')
işlem_1= ((k**x)*sym.exp(-k))
işlem_2=sym.factorial(x)
poisson_distribution=işlem_1/işlem_2 #poisson distributionun formülünü verir
pprint(poisson_distribution)
syp.plot(poisson_distribution.subs({k:2}),(x,2,10),title="Poisson Distribution")#değerleri girerek grafiği oluşturuyorum
x_value=[]
y_value=[]
for value in range(15):
    y=poisson_distribution.subs({k:2,x:value}).evalf()#k=2 olarak değer atadım
    y_value.append(y)
    x_value.append(value)
plt.plot(x_value,y_value)
```

Belirli bir nehir üzerinde sel ortalama 100 yılda meydana gelir.100 yıllık sürede maksimum 6 sel meydana gelme olasılığı nedir?

```python
x=Symbol('x')
k=Symbol('lambda')
işlem_1= ((k**x)*sym.exp(-k))
işlem_2=sym.factorial(x)
olasılık=işlem_1/işlem_2
syp.plot(olasılık.subs({k:1}),(x,0,6),title="sel olasılığı")
for i in range(7):#burada kaç selin meydana gelme olasılığının değerini hesapladım.
    sonuç=olasılık.subs({k:1,x:i})
    pprint(sonuç)
x_value=[]
y_value=[]
for value in range(7):#bu örneğin grafiğini çizdim...
    y=olasılık.subs({k:1,x:value}).evalf()#Ortalama etkinlik oranı 100 yıl başına bir taşma sel olduğu için λ = 1
    y_value.append(y)
    x_value.append(value)
plt.plot(x_value,y_value)
```
