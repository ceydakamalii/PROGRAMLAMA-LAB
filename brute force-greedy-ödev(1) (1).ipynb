```python
#Bu fonksiyon 5 kişinin random olarak okuduğu sayfa sayı aralıklarını oluşturur ve bu sayfaları bir diziye aktarır.
import random 
def list(a, b):#a=kişi sayısı,b=sayfa sayısı
    liste=[[0 for x in range(2)] for y in range(a)]#burada iki boyutlu bir dizi oluşturuyoruz
    for y in range (a):
        for x in range (2):
            liste[y][x]=random.randint(1,b)#sayfa aralıklarını random olarak atıyoruz
    print("liste başlangıç hali=",liste)
    for y in range (a): #sayfa aralıklarını random atadığımız için başlangıcı sondan küçük olarak değiştiriyoruz.
        if(liste[y][0]>liste[y][1]):
            temp=liste[y][0]
            liste[y][0]=liste[y][1]
            liste[y][1]=temp
    print("liste düzenlenmiş hali=",liste)
    liste_1=[]#boş bir dizi oluşturup okunan sayfaları atıyoruz.
    for y in range (a):
        c=liste[y][0]#başlangıç değerim(okumaya başladığım ilk sayfa)
        d=liste[y][1]#bitiş değerim(okunan son sayfa)
        while(d>=c):#son sayfadan ilk sayfaya kadar tüm elemanları listeye ekliyorum
            liste_1.append(d)#son elemanı ekliyorum listeye
            d=d-1
    print("tek boyutlu liste_1=",liste_1) 
```

```python
def frekans(liste):##Tuple kullanarak histogram yapıyorum modu bulabilmek için
    frekans_liste=[]
    for i in range(len(liste)):
        s=False
        for j in range(len(frekans_liste)):
            if(liste[i]==frekans_liste[j][0]):
                frekans_liste[j][1]=frekans_liste[j][1]+1
                s=True
        if(s==False):
            frekans_liste.append([liste[i],1])
    return frekans_liste
```

```python
import itertools
def showDifferentPages(list1):
    differentPages = [] #Bu k kişilerinin yani kesişmeyen kişilerin tutulacağı liste.
    isDifferent = True #Kesişiyorlar mı diye kontrol ifadem.
    b = 1 #Bu variable'yi sadece bir elemanı listeye atıp diğer elemanlarla karşılaştırabilmek için tanımladım.
    for i in range(len(list1)):
        list2 = []
        if (b == 1):
            list2.append(list1[i]) 
        b += 1
        for j in range(i + 1, len(list1)):
            if (list1[i][0] < list1[j][0] and list1[i][1] < list1[j][0]):
                for a in range(len(list2)):
                    if(list1[j][0] < list2[a][0] and list1[j][1] < list2[a][0]):
                        isDifferent = True

                    elif(list1[j][0] > list2[a][1] and list1[j][1] > list2[a][1]):
                        isDifferent= True
                    else:
                        isDifferent = False
                        break
                if (isDifferent == True):
                    list2.append(list1[j])
            elif (list1[i][0] > list1[j][1] and list1[i][1] > list1[j][1]):
                for a in range(len(list2)):
                    if(list1[j][0] < list2[a][0] and list1[j][1] < list2[a][0]):
                        isDifferent = True

                    elif(list1[j][0] > list2[a][1] and list1[j][1] > list2[a][1]):
                        isDifferent = True

                    else:
                        isDifferent = False
                        break
                if (isDifferent == True):
                    list2.append(list1[j])


        if (len(differentPages) == 0):
            differentPages = list2
        else:
            difMin = differentPages[0][0]
            difMax = differentPages[0][1]
            for i in differentPages:
                if (i[0] < difMin):
                    difMin = i[0]
                if (i[1] > difMax):
                    difMax = i[1]
            sizeDif = difMax - difMin

            list2Min = list2[0][0]
            list2Max = list2[0][1]
            for j in list2:
                if (j[0] < list2Min):
                    list2Min = j[0]
                if (j[1] > list2Max):
                    list2Max = j[1]
            sizeList2 = list2Max - list2Min

            if (sizeList2 > sizeDif):
                differentPages = list2
        print("en çok okuyan grup kişi sayısı",len(differentPages))
        return differentPages
```

```python
def showDifferentPagesBruteForce(list1):
    if len(list1) == 0:
        return list1
    if len(list1) == 1:
        return list1
    
    list_combinations=[]
    for i in range(2, len(list1)+1):
        x=list(itertools.combinations(list1,i))                        
        list_combinations.append(x)
    
    maxPage = 0
    for i in list1:
        sayi=i[1]-i[0]+1
        if sayi>maxPage:                                                   
            maxPage=sayi
            


    for i in list_combinations:                                         
        for k in i:                                                     
            y=list(itertools.combinations(k,2))
            exist=1
            for j in y:
                if exist == 1:
                    if ((j[0][0]<j[1][0] and j[0][1]<j[1][0]) or (j[0][0]>j[1][1] and j[0][1]>j[1][1]) ):
                        exist=1
                    else:
                        exist=0
            if exist==1:
                toplam=0
                for sayfa in k:
                    toplam=toplam+sayfa[1]-sayfa[0]+1
                if toplam>maxPage:                                          
                    maxPage=toplam
                    return k
```
